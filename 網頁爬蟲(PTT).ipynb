## 爬蟲實作: PTT八卦版範例

### 1. 建立與PTT首頁的連線

```python
# 導入模組(module) 
import requests 
# 把到ptt八卦版網址存到URL變數中
URL = "https://www.ptt.cc/bbs/Gossiping/index.html"
# 設定Header與Cookie，透過cookie才能解鎖18+的限制
my_headers = {'cookie': 'over18=1;'}
# 發送get請求到ptt八卦版
response = requests.get(URL, headers = my_headers)
# 印出回傳網頁程式碼
print(response.text)
```

### 2. 解析爬回的首頁內容

```python
# 導入模組(module)
from bs4 import BeautifulSoup
# 透過BeautifulSoup套件將送出請求得到的網頁程式碼以指定html的型態建立物件
soup = BeautifulSoup(response.text, "html.parser")
# 輸出排版後的HTML內容
print(soup.prettify())  
```

### 3. 搜尋特定內容

```python
# 透過find搜尋第一個符合的html節點
result = soup.find("div")
print(result)
```

```python
# 再以CSS屬性搜尋其節點
result = soup.find("div",class_="title")
print(result)
```

```python
# 透過find_all查找所有相符的指定標籤與節點
result = soup.find_all("div",class_="title")
print(result)
```

```python
titles = soup.find_all("div",class_="title")
# 將所有標題的連結取出的方法
for title in titles:
    print(title.select_one("a").get("href"))
# 將所有標題的文字取出的方法
for title in titles:
    print(title.select_one("a").getText())
```

```python
# 利用select即可只透過CSS進行搜尋
titles = soup.select(".title")
print(titles)
```

```python
titles = soup.find_all("div",class_="title")
# 建立一個List將所有網址存入，方便待會進行爬蟲與進入
titleList = []
for title in titles:
    try: # 可能發生因文章被刪除(連結消失)而發生爬取錯誤，因此需用try...except跳過
        print(title.select_one("a").get("href"))
        titleList.append(title.select_one("a").get("href"))
    except:
        continue
```

### 4. 進入其中一篇文章

可以透過更改titleList的編號來進入不同篇文章

```python
# 進入ptt網頁的方法，前面需加上ptt網址
ptt = "https://www.ptt.cc"
# 透過剛剛建立的List將第一個網址取出，也就是進入ptt的子網頁中
URL = ptt + titleList[15]
my_header = {'cookie': 'over18=1;'}
response = requests.get(URL, headers = my_header)
# 與爬母網頁做相同的動作，方便查閱與查找想要的資訊
soup = BeautifulSoup(response.text, "html.parser")
print(soup.prettify())  
```

```python
# 透過上述得到本頁標題名稱，以title來做查找
titleName = soup.find("title").getText()
print(titleName)
```

### 5. 資料清理

觀察規則，移除多餘的html標籤

```python
# 利用class名稱得到作者、看板、標題、與日期
headNames = soup.find_all("span",class_="article-meta-value")
headList = []
for headName in headNames:
    print(headName.getText())
    headList.append(headName.getText())
```

```python
# 替他們分別建立名稱
print("作者:"+headList[0]+"\n看板:"+headList[1]+"\n標題:"+headList[2]+"\n日期:"+headList[3])
```

```python
# 取得內文的方法(因為PTT的內文沒有class跟id只好整個body取下)
content = soup.find("div",class_="bbs-screen bbs-content")
print(content)
```

```python
content = soup.find("div",class_="bbs-screen bbs-content").getText()
# 透過split做切割，取得內文以及上半部
content = content.split("--")[0]
content = content.split("\n")
print(content)
```

```python
# 透過list將標題部分捨去，取得內文
content = content[1:]
print(content)
```

```python
# 以join加入換行部分，將內文重新排版
content = "\n".join(content)
print(content)
```

```python
# 透過特定class名稱可以得知PTT上如何得到噓與推
push = soup.find_all("div",class_="push")
print(push)
```

```python
# 利用推的class跟迴圈計算出「推」的數量
pushButtons = soup.find_all("span",class_="hl push-tag")
i = 0
for pushButton in pushButtons:
    # print(pushButton.getText())
    i += 1
print(i)
```

```python
# 利用噓的class跟迴圈計算出「噓」的數量
pushButtons = soup.find_all("span",class_="f1 hl push-tag")
i = 0
for pushButton in pushButtons:
#     print(pushButton.getText())
# 因同一個class有兩種，所以再做篩選    
    if(pushButton.getText() == "噓 "):
        i += 1
print(i)
```

### 6. 儲存成csv檔

```python
import csv
# 開啟一個新的csv檔案並寫入資料
with open('dataset/PTTcontent.csv', 'w', newline='',encoding="UTF-8") as csvfile: 
    writer = csv.writer(csvfile)
    writer.writerow(["編號", "作者", "看板","標題","日期","內文","推","噓"])
```

```python
# 接著做整合
ptt = "https://www.ptt.cc"
number = 1

for i in range(len(titleList)):
    URL = ptt + titleList[i]
    my_header = {'cookie': 'over18=1;'}
    response = requests.get(URL, headers = my_header)
    soup = BeautifulSoup(response.text, "html.parser")
    headNames = soup.find_all("span",class_="article-meta-value")
    headList = []
    for headName in headNames:
        headList.append(headName.getText())
    pushButtons = soup.find_all("span",class_="hl push-tag")
    i = 0
    for pushButton in pushButtons:
        i += 1
    headList.append(i)
    j = 0
    pushButtons = soup.find_all("span",class_="f1 hl push-tag")
    for pushButton in pushButtons:
        if(pushButton.getText() == "噓 "):
            j += 1
    headList.append(j)
    content = soup.find("div",class_="bbs-screen bbs-content").getText()
    content = content.split("--")[0]
    content = content.split("\n")
    content = content[1:]
    content = "\n".join(content)
    print(str(number)+".\n"+"作者:"+headList[0]+"\n看板:"+headList[1]+"\n標題:"+headList[2]+"\n日期:"+headList[3]+"\n內文:"+content+"\n推:"+str(headList[4])+"\n噓:"+str(headList[5]))

# 寫入資料
    with open('dataset/PTTcontent.csv', 'a', newline='',encoding="UTF-8") as csvfile: # 避免產生亂碼，需加上UTF-8統一編碼
        writer = csv.writer(csvfile)
        writer.writerow([str(number),headList[0],headList[1],headList[2],headList[3],content,headList[4],headList[5]])
    number += 1
```
